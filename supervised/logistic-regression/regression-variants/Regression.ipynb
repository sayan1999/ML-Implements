```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
plt.style.use('dark_background')
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from yellowbrick.contrib.classifier import DecisionViz
from sklearn.preprocessing import Normalizer, StandardScaler
from sklearn.metrics import classification_report
```

```python
X, y = load_iris(return_X_y=True)
xtrain, xtest, ytrain, ytest = train_test_split(X, y, random_state=23, test_size=0.2)

std=StandardScaler()
norm=Normalizer()
xtrain=std.fit_transform(xtrain)
xtrain=norm.fit_transform(xtrain)
xtest=std.transform(xtest)
xtest=norm.transform(xtest)

pca=PCA(2)
xtrain=pca.fit_transform(xtrain)
xtest=pca.transform(xtest)

plt.bar(['PCA1', 'PCA2'], pca.explained_variance_ratio_)
plt.title('Explained variance Ratio')
plt.show()
```

```python
cls=DecisionViz(
    LogisticRegression(C=0.1), title="Logistic Regression",
    features=['PCA1', 'PCA2'], classes=[0, 1, 2]
)

cls.fit(xtrain, ytrain)
cls.draw(xtrain, ytrain)
cls.show()
print('Score:', cls.score(xtrain, ytrain))
print(classification_report(ytest, cls.predict(xtest)))
```

```python
cls=DecisionViz(
    LogisticRegression(C=1), title="Logistic Regression",
    features=['PCA1', 'PCA2'], classes=[0, 1, 2]
)

cls.fit(xtrain, ytrain)
cls.draw(xtrain, ytrain)
cls.show()
print('Score:', cls.score(xtrain, ytrain))
print(classification_report(ytest, cls.predict(xtest)))
```
